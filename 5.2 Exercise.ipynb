## Riemann Sum Function

```python
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    sum_result = 0
    for k in range(1, n + 1):
        x_k = a + k * delta_x
        sum_result += f(x_k) * delta_x
    return sum_result 
```

## Using the Right Riemann Sum to Find the Area Under a Curve

### a)

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to integrate
def function_to_integrate(x):
    return np.sqrt(4 - x**2)

# Right Riemann sum function
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    sum_result = 0
    for k in range(1, n + 1):
        x_k = a + k * delta_x
        sum_result += f(x_k) * delta_x
    return sum_result

# Interval and number of rectangles
a = 0
b = 2
n = 4

# Generate x values for plotting the function
x_values = np.linspace(a, b, 100)
y_values = function_to_integrate(x_values)

# Plotting the function
plt.plot(x_values, y_values, label='$f(x) = \sqrt{4 - x^2}$')

# Plotting rectangles for the right Riemann sum R4
for k in range(1, n + 1):
    x_k = a + k * (b - a) / n
    plt.plot([x_k, x_k], [0, function_to_integrate(x_k)], 'r--')

# Adding labels and title
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Graph of $f(x) = \sqrt{4 - x^2}$ with Right Riemann Sum Rectangles')
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Function to calculate the right Riemann sum
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    sum_result = 0
    for k in range(1, n + 1):
        x_k = a + k * delta_x
        sum_result += f(x_k) * delta_x
    return sum_result

# Function representing the curve
def curve_function(x):
    return (4 - x**2)**0.5

# Interval and number of rectangles
start_point = 0
end_point = 2
rectangles = 4

# Calculate and print the right Riemann sum
riemann_sum = right_riemann_sum(curve_function, start_point, end_point, rectangles)
print("Right Riemann Sum (R4):", riemann_sum)
```

```python
# Define the function
def function_to_integrate(x):
    return np.sqrt(4 - x**2)

# Function to calculate the right Riemann sum
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    sum_result = 0
    for k in range(1, n + 1):
        x_k = a + k * delta_x
        sum_result += f(x_k) * delta_x
    return sum_result

# Interval [a, b]
a = 0
b = 2

# Calculate right Riemann sums for increasing values of n
for n in [10, 100, 1000, 10000]:
    riemann_sum_result = right_riemann_sum(function_to_integrate, a, b, n)
    print(f"Approximate Area (n={n}):", riemann_sum_result)
```

The special name for the value of the area under the curve is called the definite integral.
