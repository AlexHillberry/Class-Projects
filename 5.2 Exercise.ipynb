## Riemann Sum Function

```python
# Define the right Riemann sum function
def right_riemann_sum(f, a, b, n):
    # Calculate the width of each subinterval
    delta_x = (b - a) / n
    
    # Initialize the sum to zero
    sum_result = 0
    
    # Use a for loop to iterate over the right endpoints of subintervals
    for k in range(1, n + 1):
        # Calculate the right endpoint for the current subinterval
        x_k = a + k * delta_x
        
        # Evaluate the function at the right endpoint and multiply by the width of the subinterval
        sum_result += f(x_k) * delta_x
    
    # Return the calculated right Riemann sum
    return sum_result
```

### a)

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to integrate
def function_to_integrate(x):
    return np.sqrt(4 - x**2)

# Right Riemann sum function
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    sum_result = 0
    for k in range(1, n + 1):
        x_k = a + k * delta_x
        sum_result += f(x_k) * delta_x
    return sum_result

# Interval and number of rectangles
a = 0
b = 2
n = 4

# Generate x values for plotting the function
x_values = np.linspace(a, b, 100)
y_values = function_to_integrate(x_values)

# Plotting the function
plt.plot(x_values, y_values, label='$f(x) = \sqrt{4 - x^2}$')

# Plotting rectangles for the right Riemann sum R4 with non-overlapping lines
for k in range(1, n + 1):
    x_k = a + k * (b - a) / n
    left_edge = a + (k - 1) * (b - a) / n
    right_edge = a + k * (b - a) / n
    plt.plot([left_edge, right_edge], [function_to_integrate(x_k), function_to_integrate(x_k)], 'k-',)
    plt.plot([x_k, x_k], [0, function_to_integrate(x_k)], 'k-')
    plt.fill_between([left_edge, right_edge], 0, [function_to_integrate(x_k), function_to_integrate(x_k)], color='steelblue', alpha=0.3)

# Adding labels and title
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Graph of $f(x) = \sqrt{4 - x^2}$ with Right Riemann Sum Rectangles')
plt.grid(True)

# Display the plot
plt.show()

```
