```python
import numpy as np

# Define the coefficients matrix A and the constants vector b
A = np.array([[27, -10, 4, -29],
              [-16, 5, -2, 18],
              [-17, 4, -2, 20],
              [-7, 2, -1, 8]])

b = np.array([1, -1, 0, 1])

# Solve the system of equations using numpy
x = np.linalg.solve(A, b)

# Print
print("x1 =", x[0])
print("x2 =", x[1])
print("x3 =", x[2])
print("x4 =", x[3])
```

### b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a meshgrid
x = np.arange(-10, 11, 1)
y = np.arange(-10, 11, 1)
xx, yy = np.meshgrid(x, y)

# Define the transformation Matrix T
T = np.array([
    [0.70710678, -0.70710678],
    [0.70710678, 0.70710678]
])

# Flatten the grid
xy = np.vstack([xx.flatten(), yy.flatten()])

# Apply the linear transformation T to all points in the grid
trans = T @ xy

# Reshape 
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

# Plot original and transformed grids
f, axes = plt.subplots(1, 2, figsize=(10, 5))

# First Grid
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[0].set_title('Original Grid', color='darkblue')  # Set title color
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].axvline(0, color='black', linewidth=0.5)

# Transformation Grid
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
axes[1].set_title('Transformed Grid', color='green') 
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axvline(0, color='black', linewidth=0.5)

# Adjust layout to avoid overlapping titles
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Aesthetics
f.suptitle('Original vs. Transformed', fontsize=16)
plt.show()
```
