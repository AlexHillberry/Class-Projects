```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
file_path = r'C:\Users\Alex\OneDrive - Bellevue University\DSC 320\week2data\car_data.csv'
data = pd.read_csv(file_path)

# Set scatter color for the plot points
scatter_color = 'steelblue'

# Prepare the data
X = data[['weight']]
y = data['hwy_mpg']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Retrieve the slope (coefficient) and intercept of the best-fit line
slope_best_fit = model.coef_[0]
intercept_best_fit = model.intercept_

# Plot the scatterplot
dot_size = 15
plt.scatter(data['weight'], data['hwy_mpg'], color=scatter_color, s=dot_size, label='Graphed Values')

# Plot the manually fitted line
plt.plot(data['weight'], slope_manual * data['weight'] + intercept_manual, color='maroon', label='Line of Estimated Best-Fit')

# Plot the best-fit line
plt.plot(data['weight'], slope_best_fit * data['weight'] + intercept_best_fit, color='orange', label='Actual Best-Fit Line')
plt.title('Highway MPG vs Weight')
plt.xlabel('Weight')
plt.ylabel('Highway MPG')
plt.legend()
plt.show()
```

# 7

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Predictions for the manually fitted line
y_pred_manual = slope_manual * data['weight'] + intercept_manual

# Predictions for the best-fit line
y_pred_best_fit = model.predict(X)

# Calculate RMSE for both lines
rmse_manual = np.sqrt(mean_squared_error(y, y_pred_manual))
rmse_best_fit = np.sqrt(mean_squared_error(y, y_pred_best_fit))

print(f"RMSE for Line of Estimated Best-Fit: {rmse_manual:.2f}")
print(f"RMSE for Actual Best-Fit Line: {rmse_best_fit:.2f}")
print("Speaking generally, the difference between the two lines' RMSE is relatively small.")
print("The Actual Best-Fit Line is a better fit, being smaller than that of the estimated one.")

```

```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
file_path = r'C:\Users\Alex\OneDrive - Bellevue University\DSC 320\week2data\car_data.csv'
data = pd.read_csv(file_path)

# Set scatter color for the plot points
scatter_color = 'steelblue'

# Prepare the data
X = data[['weight']]
y = data['hwy_mpg']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Retrieve the slope and y-intercept 
slope_best_fit = model.coef_[0]
intercept_best_fit = model.intercept_

# Plot the scatterplot
dot_size = 15
plt.scatter(data['weight'], data['hwy_mpg'], color=scatter_color, s=dot_size, label='Graphed Values')

# Plot the best-fit line
plt.plot(data['weight'], slope_best_fit * data['weight'] + intercept_best_fit, color='orange', label='Actual Best-Fit Line')

# Add a prediction point and label to call in print statement
weight_to_predict = 3200
predicted_mpg = model.predict(pd.DataFrame({'weight': [weight_to_predict]}))[0]
plt.scatter(weight_to_predict, predicted_mpg, color='red', marker='X', s=100, label='Prediction Point')

plt.title('Highway MPG vs Weight')
plt.xlabel('Weight')
plt.ylabel('Highway MPG')
plt.legend()
plt.show()

# Print value of the prediction point
print(f"Predicted Highway MPG for a car weighing {weight_to_predict} pounds using the best-fit line: {predicted_mpg:.2f} mpg.")
```
