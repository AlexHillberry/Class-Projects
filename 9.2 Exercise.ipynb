### 2b)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset from the specified location
file_path = r'D:\OneDrive - Bellevue University\DSC 320\video_game_data.csv'
data = pd.read_csv(file_path)

# Calculate the covariance matrix
cov_matrix = np.cov(data[['critic_score', 'user_score']].T)

# Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort the eigenvectors by decreasing eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Scatterplot of user scores versus critic scores
plt.scatter(data['critic_score'], data['user_score'], c=data['total_shipped'], cmap='rainbow')
plt.colorbar(label='Total Shipped')
plt.title('Video Game User Scores versus Critic Scores')

# Plot first component
plt.quiver(*data[['critic_score', 'user_score']].mean(),
           *eigenvectors_sorted[:, 0] * np.sqrt(eigenvalues[0]),
           color='red', scale=1, scale_units='xy', angles='xy')

# Plot second component
plt.quiver(*data[['critic_score', 'user_score']].mean(),
           *eigenvectors_sorted[:, 1] * np.sqrt(eigenvalues[1]),
           color='blue', scale=1, scale_units='xy', angles='xy')

plt.xlabel('Critic Score')
plt.ylabel('User Score')
plt.grid(True)
plt.show()
```
