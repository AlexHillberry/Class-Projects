## 1a)

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Load qb_stats dataset
qb_stats = pd.read_csv('D:\\OneDrive - Bellevue University\\DSC 320\\qb_stats.csv')

#Find the mean of each column, suppressing the future error by specifying only numeric values.
mean = qb_stats.mean(numeric_only=True)
print(mean)
```

## 1b)

```python
# (b) Find the standard deviation of each column, suppressing the future error by specifying only numeric values.
std_values = qb_stats.std(numeric_only=True)
print(std_values)
```

## 1c)

```python
# (c) Create a histogram of the number of yards with ascending aesthetics by color.
cmap = plt.get_cmap('RdYlGn')

n, bins, patches = plt.hist(qb_stats['yds'], bins=20, edgecolor ="black")

# Assign a color to each bar in the histogram
for i, patch in enumerate(patches):
    color = cmap(i / len(patches))
    patch.set_facecolor(color)

plt.title('Number of Yards per Season', fontsize=16)
plt.xlabel('Yards')
plt.ylabel('Frequency of Yards Thrown')
plt.show()

```

## 1d)

```python
# Set colors for a football themed boxplot! =)
football_colors = dict(background='#008000', box='#8B4513', lines='black', median='white')

# Set the figure background color to green
fig, ax = plt.subplots(figsize=(5, 4))
fig.patch.set_facecolor(football_colors['background'])
ax.set_facecolor(football_colors['background'])

# Create boxplot. Add colors.
bp = ax.boxplot(qb_stats['td'], patch_artist=True, boxprops=dict(facecolor=football_colors['box'], color=football_colors['lines']),
                medianprops=dict(color=football_colors['median']))

plt.title('Boxplot of Number of Touchdowns', fontsize=16)
plt.ylabel('Frequency')

# Set the color of whiskers to black
for element in ['whiskers', 'caps']:
    plt.setp(bp[element], color=football_colors['lines'])

plt.show()

```

## 2a)

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load survey_data dataset
survey_data = pd.read_csv('D:\\OneDrive - Bellevue University\\DSC 320\\survey_data.csv')

# Find probability and set it as percent
prob_brown_hair = (survey_data['hair_color'] == 'brown').mean() * 100
print(f"Probability of having brown hair: {prob_brown_hair:.2f}%")
```

## 2b)

```python
# Find probability of students with blue eyes
prob_blue_eyes = (survey_data['eye_color'] == 'blue').mean() * 100
print(f"Probability of having blue eyes: {prob_blue_eyes:.2f}%")
```

## 2c)

```python
prob_blue_given_brown = (survey_data[survey_data['hair_color'] == 'brown']['eye_color'] == 'blue').mean() * 100
print(f"Probability of having blue eyes, given brown hair: {prob_blue_given_brown:.2f}%")
```

## 2d)

```python
prob_brown_given_blue = (survey_data[survey_data['eye_color'] == 'blue']['hair_color'] == 'brown').mean() * 100
print(f"Probability of having brown hair, given blue eyes: {prob_brown_given_blue:.2f}%")
```

## 2e)

```python
check = prob_blue_given_brown == prob_brown_hair * prob_blue_eyes
print(f"Are brown hair and blue eyes independent? {check}. "
      f"The results indicate that students having both brown hair and blue eyes are "
      f"dependent on each other due to the fact that the probability of brown hair given blue eyes is not the same as the product of "
      f"the probabilities of brown hair and blue eyes.")

```

## 2f)

```python
from matplotlib.colors import ListedColormap

# Define colors for eye colors
eye_color_colors = {'black': 'black', 'blue': 'steelblue', 'green': 'lightgreen', 'brown': 'saddlebrown', 'blonde': 'gold', 'other': 'gray'}

# Combine groups to make blonde 1 category
survey_data['hair_color'] = survey_data['hair_color'].replace({'blond': 'blonde'})

# Group by hair and eye color and plot the bar graph with specific colors for eye colors
color_counts = survey_data.groupby(['hair_color', 'eye_color']).size().unstack()
cmap = ListedColormap([eye_color_colors.get(color, 'gray') for color in color_counts.columns])
ax = color_counts.plot(kind='bar', stacked=True, colormap=cmap)

# Make legend same color for readibility
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles, labels, title='Eye Color', bbox_to_anchor=(1, 1), loc='upper left')

# Set the title and labels with capitalized text
plt.title('Hair Color and Eye Color of College Students')
plt.xlabel('Hair Color')
plt.ylabel('Count')

# Show the plot
plt.show()


```
